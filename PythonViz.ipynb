```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


#read data file into Jupyter
df = pd.read_csv('C:\\Users\\18046\\CTBA\\M2\\OpidsVA.csv')

#Set up figure and axes
fig = plt.figure()
ax = fig.add_subplot()
ax2 = ax.twinx()
width = 0.3
#Set bar chart and assign dataframe values to axes
df['MedianHouseholdIncome'].plot(kind = 'bar', color = 'grey', ax = ax, width = width, position=1)
df['FPOO-Rate'].plot(kind = 'bar', color = 'b', ax = ax2, width = width, position = 0)

#set up our legends for subplot values
ax.legend(loc = "upper left")
ax2.legend(loc = "upper right")

#assign title and labels
plt.suptitle('Household Income and Opioid Overdose Relationship in Virginia', size =20)
ax.set_xlabel('FIPSCODE', fontsize=12)
ax2.set_ylabel('Opioid Deaths per County', fontsize=16)
ax.set_ylabel('Median Income per County', fontsize=16)

#set size of our graph and show graph
fig.set_size_inches(25, 12)
plt.show()

#I created a subplot of two bar graphs, one showing household income and the other showing opiod overdoses. In a majority of the regions within the plot, as the household income decreases opiod overdoses increase. 
```

```python
#read data file into Jupyter
df = pd.read_csv('C:\\Users\\18046\\CTBA\\M2\\OpidsVA.csv')

#Set two different data series with the necessary data fields
yesSeries = pd.Series(df['FPOO-Rate'])
nodSeries = pd.Series(df['FFHO-Rate'])

#Plotting series
df = pd.DataFrame({"Opioid Overdose":yesSeries,"Heroine Overdose":nodSeries})

#Assigning plot to a bar graph, choosing color for bars
ax = df.plot(kind = 'bar', color=["Blue","Red"], figsize=(14,8))

#Setting titles and labels
plt.suptitle('Opioid and Heroine Overdose Relationship in Virginia', size =14)
ax.set_xlabel('FIPSCODE', fontsize=10)
ax.set_ylabel('Deaths per County', fontsize=10)

#show graph
plt.show()

#I created a double bar graph that tracks frequency of deaths between opioid and heroine overdoses. There seems to be no strong consistency in the results which leads me to believe opioid use does not lead to heroine use.
```

```python
#read data file into Jupyter
df = pd.read_csv('C:\\Users\\18046\\CTBA\\M2\\presidents.csv')

#Assigned data field to a variable and plotted it
data = df['State']
ax = data.value_counts().plot.bar(figsize=(14,8), color = 'red')

#Setting title, and axis labels
plt.suptitle('Number of Presidents Born in Each State', size =14)
ax.set_xlabel('State', fontsize=10)
ax.set_ylabel('Number of Presidents Born', fontsize=10)

#I set up a bar plot that counts the frequency of presidential births per state; Virginia has the most.
```
