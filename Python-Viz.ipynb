```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#read data file into Jupyter
df = pd.read_csv('C:\\Users\\18046\\CTBA\\M2\\OpidsVA.csv')


#Set up figure and axes
fig = plt.figure()
ax = fig.add_subplot()
ax2 = ax.twinx()
width = 0.3
#Set bar chart and assign dataframe values to axes
df['MedianHouseholdIncome'].plot(kind = 'bar', color = 'grey', ax = ax, width = width, position=1)
df['FPOO-Rate'].plot(kind = 'bar', color = 'b', ax = ax2, width = width, position = 0)

#set up our legends for subplot values
ax.legend(loc = "upper left")
ax2.legend(loc = "upper right")

#assign title and labels
plt.suptitle('Household Income and Opioid Overdose Relationship in Virginia', size =20)
ax.set_xlabel('FIPSCODE', fontsize=12)
ax2.set_ylabel('Opioid Deaths per County', fontsize=16)
ax.set_ylabel('Median Income per County', fontsize=16)

#set size of our graph and show graph
fig.set_size_inches(25, 12)
plt.show()

#I created a subplot of two bar graphs, one showing household income and the other showing opiod overdoses. In a majority of the regions within the plot, as the household income decreases opiod overdoses increase. 
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#read data file into Jupyter
df = pd.read_csv('C:\\Users\\18046\\CTBA\\M3\\ConstructionTimeSeriesDataV2.csv')

#Set subplots and rename columns
fig, ax = plt.subplots()
df.rename(columns = {'Total Construction': 'Total', 'Private Construction': 'Private', 'Public Construction': 'Public'}, inplace = True)
x = df['Month-Year']

#Plot all three columns
ax.plot(df['Month-Year'], df['Total'], color = 'g', linewidth =2, label ='Total Spending')
ax.plot(df['Month-Year'], df['Private'], color = 'b', linewidth =.5, alpha = 0.8, label ='Private Spending')
ax.plot(df['Month-Year'], df['Public'], color = 'r', linewidth =.5, alpha = 0.8, label ='Public Spending')


#Hide top and right side spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Setting labels
ax.set_xlabel('Year', fontsize = '15')
ax.set_ylabel('Cost', fontsize = '15')
ax.set_xticks(x[::12])
#Setting 'mean' lines and labeling them
ax.axhline(y = 79470, linewidth = 0.8, color = 'k', linestyle = '--', alpha = 0.5)
ax.axhline(y = 57372, linewidth = 0.8, color = 'k', linestyle = '--', alpha = 0.5)
ax.axhline(y = 22098, linewidth = 0.8, color = 'k', linestyle = '--', alpha = 0.5)
ax.text(146,81000,'Total Construction Mean',color='k', fontsize = '9')
ax.text(146,59000,'Private Construction Mean',color='k', fontsize='9')
ax.text(146,24000,'Public Construction Mean',color='k', fontsize = '9')

#Creating axis captions
capx = 'Years which construction iniatives occured'
fig.text(0.5,0, capx, ha='center', fontsize=11, bbox={"facecolor":"grey", "alpha":0.5, "pad":5})

capy = 'Total spending on construction'
fig.text(0.03,0.31, capy, fontsize=11, rotation = 90, ha='center', bbox={"facecolor":"grey", "alpha":0.5, "pad":5})

#Setting title, plot size and legend
plt.suptitle('Total Construction Budget Makeup', fontsize=18)
fig.set_size_inches(14,6)
ax.legend(loc = 'upper left')

#The insight I would use this graph to visualize during a presentation would be that total construction cost is primarily influenced by the spending on private construction as public construction costs are fairly constant.
```
